### 이미지 시각화

#### package 불러오기

```python
import numpy as np

from PIL import Image

import matplotlib.pyplot as plt

%matplotlib inline
```

#### 이미지 파일 열기

```python
path = 'source/dog.jpeg'

image_pil = Image.open(path)
image = np.array(image_pil)
```

```python
image.shape #세로길이, 가로길이, RGB
```

#### 이미지 들여다 보기

```python
image.shape
```

```python
np.min(image), np.max(image)
```

이미지를 열기 전에 shape및 min,max를 통해서 이미지의 range확인 필요

#### 그래프로 시각화 하기

```python
plt.hist(image.ravel(), 256, [0,256])
plt.show()
```

#### 그림 나타내기

```python
plt.imshow(image)
plt.show()
```
