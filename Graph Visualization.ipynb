### Graph Visualization

### Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# 주피터 내부에 그림을 띄우겠다는 명령어
```

### Basic Attributes
- alpha : 투명도
- kind : 그래프 종류 "line","bar","barh","kde"
- logy : Y축에 대해 Log scaling
- use_index : 객체의 색인을 눈금 이름으로 사용할지 여부
- rot : 눈금 이름 돌리기(rotating) 0~360
- xticks,yticks : x,y축으로 사용할 값
- xlim,ylim : X,Y축의 한계
- grid : 축의 그리드를 표현할지 여부
- subplots : 각 column에 독립된 subplot그리기
- sharex,sharey : subplots=True이면 같은 X,Y축을 공유하고 눈금과 한계를 연결
- figsize : 생성될 그래프의 크기를 tuple로 지정
- title : 그래프의 제목 지정
- legend : subplot의 범례 지정
- sort_columns : column을 알파벳 순서로 그린다

### Matplotlib 사용하기

#### 점선 그래프 그리기

```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

#### 여러 그래프 그릴 준비 하기

```python
plt.subplot(1,2,1)#행 개수, 열 개수, 순서
plt.subplot(1,2,2)
plt.show()
```

#### Multi Graph 그리기

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)

plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)

plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))

plt.show()
#순서가 중요함!
```

#### 그래프 선 옵션
- 그래프를 그릴 때 표시되는 색이나 마커 패턴을 바꾸는 것 확인
 - 색상 : b(파란색), g(초록색), r(빨간색), c(청록색), y(노란색), k(검은색), w(흰색)
 - 마커 : o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)

```python
plt.plot(data,'gv')
plt.show()
```

```python
plt.plot(data,'k*')
plt.show()
```

### 그래프 사이즈 조절
- plt.figure 안ㅇ[ figsize를 이용하여 가로, 세로 길이 조정 가능(inch 단위)


```python
plt.figure(figsize=(10,5))
plt.plot(data,'k+')
plt.show()
```

- 여러 그래프 그리고 그에 대한 크기 조절

```python
plt.figure(figsize=(1,1))
#figure는 subplot보다 위에 있어야 함
#show까지의 그래프가 적용됨
plt.subplot(2,2,1)
plt.plot(data)

plt.subplot(2,2,2)
plt.hist(hist_data, bins = 20)

plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))

plt.show()
```

- 비율 조절

#### 그래프 겹치기 + legend 달기

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'y-', drawstyle='steps-post', label='steps-post')
plt.legend() #label보이게
plt.show()
#subplot을 해주지 않고 show까지 가면 겹쳐서 나옴
```

#### 이름 달기

```python
plt.plot(np.random.randn(1000).cumsum())
plt.title('Random Graph')
plt.xlabel('Stages')
plt.ylabel('Values')
plt.show()
```

#### 종합

```python
plt.title('Graph')
plt.plot(np.random.randn(1000).cumsum(),
        'k', label='one')
plt.plot(np.random.randn(1000).cumsum(),
        'b', label='two')
plt.plot(np.random.randn(1000).cumsum(),
        'r', label='three')
plt.legend()
plt.show()
```
